# After running rf

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

oX = original_data["x"]
oy = original_data["y"]

store_max = {}

# y has to be 1D because right now it includes index
if len(oy.shape) == 2:
    oy = oy.iloc[:, 0]
oy = oy.astype(int)

real_y0 = oX.loc[oy == 0]
real_y1 = oX.loc[oy == 1]
syn_y0 = cvae_y0
syn_y1 = cvae_y1

for col in oX.columns:
    q99 = oX[col].quantile(0.99)
    max_val = oX[col].max()
    store_max[col] = max_val
    # max approximately equals 99th percentile
    if abs(max_val - q99) < 1e-6: 
        print(col, max_val)

```

```python
from matplotlib import pyplot as plt


def summarize_feature(real: pd.Series, syn: pd.Series, ceiling=None, name=""):
    qs = [0.5, 0.9, 0.95, 0.99]
    out = {
        "feature": name,
        "real_mean": real.mean(),
        "syn_mean": syn.mean(),
        "real_std": real.std(),
        "syn_std": syn.std(),
    }
    for q in qs:
        out[f"real_q{int(q*100)}"] = real.quantile(q)
        out[f"syn_q{int(q*100)}"] = syn.quantile(q)
    if ceiling is not None:
        out["real_p_at_ceiling"] = (real == ceiling).mean()
        out["syn_p_at_ceiling"] = (syn == ceiling).mean()
    return out

import numpy as np

def plot_feature(real, syn, title="", bins=60, ceiling=None):
    plt.figure()

    all_vals = np.r_[real.values, syn.values]
    edges = np.histogram_bin_edges(all_vals, bins=bins)

    plt.hist(real.values, bins=edges, alpha=0.5, density=True, label="real")
    plt.hist(syn.values,  bins=edges, alpha=0.5, density=True, label="synthetic")
    if ceiling is not None:
        plt.axvline(ceiling, linestyle="--")

    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Denisties")
    plt.legend()
    plt.show()

feat = "V10_blood_IgGRBD"
CEILING = store_max[feat] 

# y=0
summ_y0 = summarize_feature(real_y0[feat], syn_y0[feat], ceiling=CEILING, name=f"{feat} (y=0)")
# y=1
summ_y1 = summarize_feature(real_y1[feat], syn_y1[feat], ceiling=CEILING, name=f"{feat} (y=1)")

display(pd.DataFrame([summ_y0, summ_y1]))

plot_feature(
    real_y0[feat],
    syn_y0[feat],
    title=f"{feat} Distribution (y=0)",
    ceiling=CEILING
)

plot_feature(
    real_y1[feat],
    syn_y1[feat],
    title=f"{feat} Distribution (y=1)",
    ceiling=CEILING
)
```

```python

def ensure_df(X, columns):
    """Make sure X is a DataFrame with the given columns."""
    if isinstance(X, pd.DataFrame):
        return X[columns].copy()
    return pd.DataFrame(X, columns=columns)

def rf_auc_probe_binary(real_df: pd.DataFrame, syn_df: pd.DataFrame, seed=42, n_splits=5):
    """
    RF probe: label s=0 real, s=1 synthetic.
    Returns mean/std AUC via stratified CV.
    """
    n = min(len(real_df), len(syn_df))
    rng = np.random.default_rng(seed)

    # sample same n from each
    real_idx = rng.choice(real_df.index.to_numpy(), size=n, replace=False)
    syn_idx  = rng.choice(syn_df.index.to_numpy(),  size=n, replace=False)

    real_s = real_df.loc[real_idx]
    syn_s  = syn_df.loc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0).to_numpy()
    s = np.r_[np.zeros(n, dtype=int), np.ones(n, dtype=int)]

    rf = RandomForestClassifier(
        n_estimators=500,
        max_depth=None,
        random_state=seed,
        n_jobs=-1,
    )

    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    aucs = cross_val_score(rf, X, s, cv=cv, scoring="roc_auc")

    return float(aucs.mean()), float(aucs.std()), aucs

# Per-stratum probes 

cols = oX.columns

syn_y0_df = ensure_df(syn_y0, cols)
syn_y1_df = ensure_df(syn_y1, cols)

real_y0_df = ensure_df(real_y0, cols)
real_y1_df = ensure_df(real_y1, cols)

auc0_mean, auc0_std, auc0_folds = rf_auc_probe_binary(real_y0_df, syn_y0_df, seed=42)
auc1_mean, auc1_std, auc1_folds = rf_auc_probe_binary(real_y1_df, syn_y1_df, seed=42)

print(f"RF AUC (y=0): {auc0_mean:.3f} ± {auc0_std:.3f}   folds={np.round(auc0_folds,3)}")
print(f"RF AUC (y=1): {auc1_mean:.3f} ± {auc1_std:.3f}   folds={np.round(auc1_folds,3)}")
```

# Ablation section

```python
real_X = oX
real_y = oy

syn_X = pd.concat([syn_y0, syn_y1], axis=0, ignore_index=True)
syn_y = np.r_[
    np.zeros(len(syn_y0), dtype=int),
    np.ones(len(syn_y1), dtype=int)
]

# Balanced counts for probe
n_neg = min((real_y == 0).sum(), (syn_y == 0).sum())
n_pos = min((real_y == 1).sum(), (syn_y == 1).sum())
```

```python


rng = np.random.default_rng(42)

def strat_samp(neg_i, posi_i, n_neg, n_posi):
    chosen_neg = rng.choice(neg_i, size=n_neg, replace=False)
    chosen_posi = rng.choice(posi_i, size=n_posi, replace=False)
    return np.concatenate([chosen_neg, chosen_posi])

def rf_auc_probe_stratified(real_X: pd.DataFrame, real_y,
                             syn_X: pd.DataFrame, syn_y,
                             n_neg, n_pos, seed=42):

    real_y = np.asarray(real_y).astype(int)
    syn_y  = np.asarray(syn_y).astype(int)

    # indices per stratum
    real_neg_i = np.where(real_y == 0)[0]
    real_pos_i = np.where(real_y == 1)[0]

    syn_neg_i = np.where(syn_y == 0)[0]
    syn_pos_i = np.where(syn_y == 1)[0]

    # stratified sampling
    real_idx = strat_samp(real_neg_i, real_pos_i, n_neg, n_pos)
    syn_idx  = strat_samp(syn_neg_i,  syn_pos_i,  n_neg, n_pos)

    real_s = real_X.iloc[real_idx]
    syn_s  = syn_X.iloc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0).to_numpy()
    s = np.r_[np.zeros(len(real_s), dtype=int),
              np.ones(len(syn_s), dtype=int)]

    rf = RandomForestClassifier(
        n_estimators=500,
        random_state=seed,
        n_jobs=-1
    )

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    aucs = cross_val_score(rf, X, s, cv=cv, scoring="roc_auc")

    return float(aucs.mean()), float(aucs.std()), aucs
```

# Ablation function

```python
def run_ablation(real_X, real_y, syn_X, syn_y, drop_feats, n_neg, n_pos, seed=42, label=""):
    # keep only features that exist
    drop_feats = [f for f in drop_feats if f in real_X.columns]

    real_ab = real_X.drop(columns=drop_feats)
    syn_ab  = syn_X.drop(columns=drop_feats)

    auc_mean, auc_std, folds = rf_auc_probe_stratified(
        real_X=real_ab,
        real_y=real_y,
        syn_X=syn_ab,
        syn_y=syn_y,
        n_neg=n_neg,
        n_pos=n_pos,
        seed=seed
    )

    print(f"[{label}] dropped {len(drop_feats)} features")
    print(f"  AUC: {auc_mean:.3f} ± {auc_std:.3f}  folds={np.round(folds,3)}")

    return auc_mean, auc_std, folds, drop_feats
```

# RF-driven ablation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

def fit_rf_get_importance(real_X, real_y, syn_X, syn_y, n_neg, n_pos, seed=42):
    real_y = np.asarray(real_y).astype(int)
    syn_y  = np.asarray(syn_y).astype(int)

    real_neg_i = np.where(real_y == 0)[0]
    real_pos_i = np.where(real_y == 1)[0]
    syn_neg_i  = np.where(syn_y == 0)[0]
    syn_pos_i  = np.where(syn_y == 1)[0]

    real_idx = strat_samp(real_neg_i, real_pos_i, n_neg, n_pos)
    syn_idx  = strat_samp(syn_neg_i,  syn_pos_i,  n_neg, n_pos)

    real_s = real_X.iloc[real_idx]
    syn_s  = syn_X.iloc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0)
    s = np.r_[np.zeros(len(real_s), dtype=int), np.ones(len(syn_s), dtype=int)]

    rf = RandomForestClassifier(n_estimators=800, random_state=seed, n_jobs=-1)
    rf.fit(X, s)

    imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
    return imp

imp = fit_rf_get_importance(real_X, real_y, syn_X, syn_y, n_neg, n_pos)
imp.head(20)
```

```python
for k in [1, 2, 5, 10, 20, 30, 60]:
    drop_k = imp.head(k).index.tolist()
    auc_k, sd_k, folds_k, _ = run_ablation(real_X, real_y, syn_X, syn_y, drop_k, n_neg, n_pos, label=f"drop top {k}")
```
